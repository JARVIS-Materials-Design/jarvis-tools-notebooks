```python
from jarvis.db.jsonutils import loadjson
import glob
mem=[]
ids=[]
for i in glob.glob("*chgnet/*results.json"):
  d=loadjson(i)
  mem.append(d)
  ids.append(i.split('_')[0])
```

```python
import pandas as pd
df=pd.DataFrame(mem)
df['id']=ids
df.head()
```

```python
initial_a=[i['energy']['initial_a'] for i in mem]
final_a=[i['energy']['final_a'] for i in mem]
```

```python
import plotly.express as px
fig = px.scatter(x=initial_a, y=final_a)
fig.show()
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(initial_a,final_a)
```
