```python
from jarvis.db.figshare import data
```

```python
fls = data("raw_files")
```

```python
fls.keys()
```

```python
wann=fls['WANN']
```

```python
wann[0]
```

```python
dft_3d = data('dft_3d')
```

```python
import pandas as pd
df_dft3d = pd.DataFrame(dft_3d)
df_wtbh = pd.DataFrame(wann)
```

```python
df_wtbh['jid']=df_wtbh['name'].apply(lambda x: x.split('.zip')[0])
```

```python
df_wtbh
```

```python
df_common = pd.merge(df_wtbh,df_dft3d,on='jid')
```

```python
df_common.columns
```

```python
df_common_mag = df_common[(df_common['magmom_oszicar']!='na')]
```

```python
df_common_mag_spil = df_common_mag[(df_common_mag['magmom_oszicar']>0.5) & (df_common_mag['spillage']!='na') ]
```
