Get data from JARVIS-DFT OPTIMADE API

See more example: https://jarvis.nist.gov/optimade/jarvisdft/

```python
import pprint
import requests
url = "https://jarvis.nist.gov/optimade/jarvisdft/v1/structures/?filter=elements HAS  ALL C,Si"
while True:
  if url is not None:
    # print('url', url)
    response = requests.get(url)
    data = response.json()['data']
    for i in data:
      print(i)
    url=response.json()['links']['next']
  else:
    break
```
