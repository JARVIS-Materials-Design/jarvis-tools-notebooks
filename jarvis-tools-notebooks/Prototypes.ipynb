```python
from jarvis.db.figshare import data

dft=data('dft_3d')
```

```python
import pandas as pd
df=pd.DataFrame(dft)
```

```python
from jarvis.core.composition import Composition
Composition.from_string('Al2O3').prototype.replace('.0','')
```

```python
from tqdm import tqdm
tqdm.pandas()
from jarvis.core.composition import Composition
import string
def get_proto(formula=''):
    comp=Composition.from_string(formula)
    reduced, repeat = comp.reduce()
    proto = ""
    all_upper = string.ascii_uppercase
    items = sorted(list(reduced.values()), reverse=True)
    N = 0
    # for specie, count in reduced.items():
    for c in items:
        proto = proto + str(all_upper[N]) + str(int(c))
        N = N + 1

    return proto

df['prototype']=df['formula'].progress_apply(lambda x:get_proto(x))
```

```python
df['prototype_spg']=df['prototype']+'_'+df['spg_number']
```
