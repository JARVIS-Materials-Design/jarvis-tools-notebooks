```python
from jarvis.db.figshare import data
import pandas as pd
d=data('dft_3d')
df=pd.DataFrame(d)
```

Example: find La2O3

```python
df[df['formula']=='La2O3']
```

Example: find MnO2

```python
df[df['formula']=='MnO2']
```
