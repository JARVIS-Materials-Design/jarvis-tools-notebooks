```python
from jarvis.db.figshare import data
import pandas as pd
from jarvis.analysis.elastic.tensor import ElasticTensor
d=data('dft_3d')
df=pd.DataFrame(d)
dff=df[df['bulk_modulus_kv']!='na']
dff['calc_kv']=dff['elastic_tensor'].apply(lambda x: ElasticTensor(x).voigt_modulus[0])
```

```python
import plotly.express as px
fig = px.scatter(dff,x='bulk_modulus_kv', y='calc_kv',hover_data=['jid'])
fig.show()
```
