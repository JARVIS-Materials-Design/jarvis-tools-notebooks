```python
from jarvis.db.figshare import data

d=data('vacancydb')
```

```python
import os
from jarvis.core.atoms import Atoms
os.chdir('/content')

if not os.path.exists('DataDir'):
   os.makedirs('DataDir')
os.chdir('DataDir')
f=open('id_prop.csv','w')
for i in d:
  id=str(i['id'])
  atoms=Atoms.from_dict(i['bulk_atoms'])
  atoms.write_poscar(id)
  e_vac=i['ef']
  line=str(id)+','+str(e_vac)+'\n'
  f.write(line)
f.close()

os.chdir('/content')
```
