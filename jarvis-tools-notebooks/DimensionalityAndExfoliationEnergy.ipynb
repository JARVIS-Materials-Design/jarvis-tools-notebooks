```
from jarvis.db.figshare import data
from collections import Counter
import pandas as pd
import numpy as np
dft_3d = data('dft_3d')
df=pd.DataFrame(dft_3d)
threed=[]
twod=[]
oned=[]
zerod=[]
dims=Counter(df['dimensionality'].replace('na',np.nan).dropna().values)
```

```
import plotly.graph_objects as go

fig = go.Figure(go.Pie(
    name = "",
    values = list(dims.values()),
    labels = list(dims.keys()),
    hovertemplate = "%{label}: <br>Percent.: %{percent} </br> %{text}"
))

fig.show()
```
