#3D dataset

```python
from jarvis.db.figshare import data
import pandas as pd
dft3d = pd.DataFrame(data('dft_3d'))
```

```python
common_3d = (dft3d[dft3d['reference'].apply(lambda x: ('mp' or 'mvc') in x)])
```

```python
common_3d
```

```python
common_3d['reference'].unique().shape
```

```python
urls=[]
for ii, i in common_3d.iterrows():
  mp_url = 'https://materialsproject.org/materials/'+str(i['reference'])
  if mp_url not in urls:
    urls.append(mp_url)
    jid_url = 'https://www.ctcms.nist.gov/~knc6/static/JARVIS-DFT/'+str(i['jid'])+'.xml'
    print ('mp_url', mp_url)
    print ('jid_url', jid_url)
    if ii>10:
        break
```
